```python

def get_kaggle_dataset():
    """pick a random dataset from kaggle - Does NOT download data (yet)"""

    # connect to kaggle
    api = KaggleApi()
    api.authenticate()
    
    # get info about all the datasets
    datasets = api.dataset_list()

    # randomly pick one
    # random.choice picks uniformly at random. if there are N total datasets, the probability of picking any one particular dataset is 1/N
    dataset = random.choice(datasets)
    
    # print basic information
    print(f"Selected dataset: {dataset.title}")
    print(f"Description: {dataset.subtitle}")

    return dataset
    
    # download
    # api.dataset_download_files(dataset.ref, path=download_path, unzip=True)
```

```python
# web scraping
# there are easy to scrape websites and hard to scrape websites
import requests # this is to make http requests (let me read this page, for any page on the internet)
from bs4 import BeautifulSoup
```

```python
# https://en.wikipedia.org/wiki/2020_Summer_Olympics

response = requests.get("https://en.wikipedia.org/wiki/2020_Summer_Olympics")
response.status_code
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
# Extract paragraphs
paragraphs = []
for p in soup.find_all("p"):
    paragraphs.append(p.get_text())

# Extract images
images = []
for img in soup.find_all("img"):
    src = img.get("src")
    alt = img.get("alt")
    images.append({"src": src, "alt": alt})

# Extract tables
tables = []
for table in soup.find_all("table"):
    headers = []
    for th in table.find_all("th"):
        headers.append(th.get_text())

    rows = []
    for row in table.find_all("tr"):
        cells = []
        for td in row.find_all("td"):
            cells.append(td.get_text())
        rows.append(cells)

    tables.append({"headers": headers, "rows": rows})
```
